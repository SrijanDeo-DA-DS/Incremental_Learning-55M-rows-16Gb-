```python
# Compute the number of rows
num_rows = df.shape[0]

num_columns = len(df.columns)

# Get the shape as a tuple
shape = (num_rows, num_columns)
print(shape)
```

```python
df = df.drop(columns=['customer_ID','S_2'])
```

```python
# Get the data types of each column
dtypes = df.dtypes

# Print the data types
print(dtypes)
```

```python
# Define a function to fill missing values based on dtype
def fill_missing_values(df):
    # Separate columns by dtype
    float_cols = dtypes[dtypes == 'float64'].index
    object_cols = dtypes[dtypes == 'string[pyarrow]'].index

    # Compute the median for float columns
    medians = df[float_cols].median()
    
    # Compute the mode for object columns
    modes = df[object_cols].mode().iloc[0]

    # Fill float columns with median
    for col in float_cols:
        df[col] = df[col].fillna(medians[col])
    
    # Fill object columns with mode
    for col in object_cols:
        df[col] = df[col].fillna(modes[col])
    
    return df

# Apply the function to fill missing values
df = fill_missing_values(df)
```

```python
missing_values = df.isnull().sum()

percentage_missing = (missing_values / num_rows) * 100

#Print the result
print(percentage_missing)
```
